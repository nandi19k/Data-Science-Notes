# Missing Values

```python
import numpy as np 
import pandas as pd

# Create a dataframe
df = pd.DataFrame({'A':[1,2,np.nan], 'B':[6,np.nan,np.nan], 'C':[1,2,3]}) 
df
```

```python
# Check missing values in each column
df.isna().sum()
```

```python
# Remove rows with NaN values
df.dropna()
```

```python
# Original dataframe remains unchanged
df
```

```python
# Remove columns with NaN values
df.dropna(axis=1)
```

```python
# Fill NaN values with a specific value
df.fillna(value = 'Mean or median or mode of data')
```

```python
# Fill NaN values with the mean of the column
df['A'].fillna(value = df['A'].mean())
```

# Groupby

```python
import pandas as pd

# Create a dataframe
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'], 
        'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'], 
        'Sales':[200,120,340,124,243,350]} 
df = pd.DataFrame(data)
```

```python
# Group by Company (returns a GroupBy object)
df.groupby('Company')
```

```python
# Store the GroupBy object
by_comp = df.groupby('Company')
```

```python
# Minimum values in each group
by_comp.min()
```

```python
# Maximum values in each group
by_comp.max()
```

```python
# Count of values in each group
by_comp.count()
```

```python
# Detailed statistics of each group
by_comp.describe()
```

# combining Dataframes

```python
import pandas as pd

# Create example dataframes
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3'], 
                    'C': ['C0', 'C1', 'C2', 'C3'], 
                    'D': ['D0', 'D1', 'D2', 'D3']}, 
                   index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'], 
                    'B': ['B4', 'B5', 'B6', 'B7'], 
                    'C': ['C4', 'C5', 'C6', 'C7'], 
                    'D': ['D4', 'D5', 'D6', 'D7']}, 
                   index=[4, 5, 6, 7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'], 
                    'B': ['B8', 'B9', 'B10', 'B11'], 
                    'C': ['C8', 'C9', 'C10', 'C11'], 
                    'D': ['D8', 'D9', 'D10', 'D11']}, 
                   index=[8, 9, 10, 11])
```

```python
# Concatenate vertically (row-wise)
pd.concat([df1, df2, df3])
```

```python
# Concatenate horizontally (column-wise)
pd.concat([df1, df2, df3], axis=1)
```

```python
# Concatenate horizontally (column-wise)
pd.concat([df1, df2, df3], axis=1)
```

# Pandas Operations

```python
import pandas as pd

# Create example dataframe
df = pd.DataFrame({'Col1':[1,2,3,4], 
                   'Col2':[44,55,66,55], 
                   'Col3':['a','b','c','d']})
```

```python
# View first rows
df.head()
```

```python
# View last rows
df.tail()
```

```python
# Get unique values in a column
df['Col2'].unique()
```

```python
# Count unique values
df['Col2'].nunique()
```

```python
# Count frequency of each value
df['Col2'].value_counts()
```

```python
# Define a function to apply to a column
def square(x): 
    return x*x
```

```python
# Apply function to column
df['Col1'] = df['Col1'].apply(square)
```

```python
# Sum a column
df['Col1'].sum()
```
